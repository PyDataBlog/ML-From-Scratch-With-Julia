```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
from urllib.request import urlretrieve
```

```python
url = "https://raw.githubusercontent.com/PyDataBlog/NN-From-Scratch-With-Julia/master/train_catvnoncat.h5"
```

```python
urlretrieve(url, "python_train.h5")
```

```python
def load_data():
    '''
        Load some toy data
    '''
    # read file
    train_dataset = h5py.File('python_train.h5', "r")
    
    # read training array data
    train_set_x_orig = np.array(train_dataset["train_set_x"]) # shape = (209, 64, 64, 3)
    
    # flatten array data
    train_set_x_flat = train_set_x_orig.reshape(train_set_x_orig.shape[0], -1).T # shape = (12288, 209)
    
    return train_set_x_orig, train_set_x_flat
```

```python
train_x_orig, train_x_flat = load_data()
```
